Best way to form trendlines?
Let us just assume we want to create trendlines along the lows of a data series. one way of doing this would be to connect all possible lows up to time t. A trendline could then be defined as a line that touches an arbitrary number of lows; say 3. To build such a strategy we must first define a low and also define "touch". 

```python
import yfinance as yf
import scipy.signal as sc
import matplotlib.pyplot as plt
import pandas as pd
from findpeaks import findpeaks
from caerus import caerus
import trendlines as tl
import talib as ta
```

```python
sp = yf.download("^SPX")["Close"]
```

```python
hl = tl.get_highs_lows(sp)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(sp, color = "blue")
ax.scatter(hl["highs"].index, hl["highs"] , marker = "v", color = "green")
ax.scatter(hl["lows"].index, hl["lows"] , marker = "^", color = "red")
```
